```python
import pandas as pd
import os
import numpy as np
from matplotlib.finance import candlestick2_ohlc
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
%matplotlib inline
```

```python
#list content in folder, if csv
files = [x for x in os.listdir('data') if x.endswith('.csv')]
for filename in files:
    print(filename)
```

```python
df = pd.read_csv('data/AAPL.csv')
df.head()
```

```python
df.shape
```

```python
df2 = df[df.index > 9200].copy()
```

```python
#make plot bigger
plt.rcParams['figure.figsize'] = (20,10)

ax = plt.subplot(1,1,1)
candlestick2_ohlc(ax, df2['Open'], df2['High'], df2['Low'], df2['Close'], width=2, 
                  colorup='g', colordown='r', alpha=1)
plt.show()
```

```python
df.plot(kind='line', x='Date', figsize=(20,20))
```

```python
#Open 	High 	Low 	Close 	Adj Close 	Volume
df.plot(kind='line', x='Date', y=['Open', 'High', 'Low', 'Close', 'Adj Close'], figsize=(20,20))
```
